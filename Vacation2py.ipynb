```python
import gmaps
import pandas as pd
import os
import csv
from config import g_key
import requests
```

```python
gmaps.configure(api_key=g_key)
```

```python
city_csv = os.path.join('..', 'python-api-challenge', 'city_file.csv')
city_df = pd.read_csv(city_csv)
city_df
```

```python
locations = city_df[["Latitude","Longitude"]]
humidity = city_df["Humidity"].astype(float)
wind = city_df["Wind"]
temp = city_df["Temperature"].astype(float)
```

```python
fig = gmaps.figure(map_type='HYBRID')
```

```python
heat_layer = gmaps.heatmap_layer(locations, weights=humidity, dissipating=False, max_intensity=10, point_radius=1)
```

```python
fig.add_layer(heat_layer)
```

```python

city_coords = ["-16.23, 39.91", "-21.21, -159.78", "31.98, 35.69", "33.99, 50.68", "40.43, 71.72", "31.50, 34.37"]

hotels = []

# rewrite params dict
for coords in city_coords:
    target_type = "lodging"
    radius = 5000
    params = {
        "location": coords,
        "types": target_type,
        "radius": radius,
        "key": g_key
    }
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    # Run request
    response = requests.get(base_url, params)
    city_hotel = response.json()
    hotels.append(city_hotel["results"][0]["name"])
    print(city_hotel['results'][0]['name'])
    
```

```python
marker_locations = [(-16.23, 39.91), (-21.21, -159.78), (31.98, 35.69), (33.99, 50.68), (40.43, 71.72), (31.50, 34.37)]
fig = gmaps.figure()
markers = gmaps.marker_layer(marker_locations)
fig.add_layer(markers)
fig.add_layer(heat_layer)
fig
```
